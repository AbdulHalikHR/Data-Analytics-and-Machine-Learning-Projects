Machine Learning Project 1: **Analyze Your Own Expenses**

**Load Data in Collab**

```python
import pandas as pd
df = pd.read_csv("OwnExpenses.csv")

print(df.head())

#Displays header values


```

**Clean and Filter Data**

```python


# Removes missing values
df.dropna(inplace = True)

```

```python
# Fixing category names
df['Category'] = df['Category'].str.capitalize()
```

**Analyze Your Spening**

```python
# Findout Total Spending

total_spent = df['Amount'].sum()
print("Total Spent is: ", total_spent,"INR")
```

```python
# Finding the most spent Category

category_spent = df.groupby("Category")["Amount"].sum()
print("Amount Spent by categorical wise: ", category_spent,"INR")
max_spent = category_spent.idxmax()
max_spent_amount = category_spent.max()
print("Maximum amount spent is ", max_spent,":",max_spent_amount,"INR")
```

**Visualize your Expenses:**

```python
# Bar Chart - Spending/Category

import matplotlib.pyplot as plt

category_spent.plot(kind = "bar", color = "skyblue")
plt.xlabel("Category")
plt.ylabel("Amount Spent")
plt.title("Amount Spent by Category")
plt.show()
```

```python
# Pie Chart - Spending Distribution
category_spent.plot(kind = "pie", autopct = "%1.1f%%")
plt.title("Spending Distribution")
plt.ylabel("")
plt.show()
```
